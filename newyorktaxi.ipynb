```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
car_df=pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Datasets/main/TaxiFare.csv")
```

```python
car_df.shape
```

```python
car_df.sample(10)
```

```python
car_df.drop("latitude_of_dropoff",axis=1,inplace=True)
```

```python
car_df['no_of_passenger']=car_df['no_of_passenger'].replace({1:'america',2:'europe',3:'asia',4:'britain', 5:'london'})
car_df.sample(10)
```

```python
#one hot  encoding
car_df=pd.get_dummies(car_df,columns=['no_of_passenger'])
car_df.sample(10)
```

```python
car_df.isnull().sum()
```

```python
car_df.describe()
```

```python
car_df.dtypes
```

```python
sns.pairplot(car_df,diag_kind="kde")#bivariate Plots
```

```python
car_df['pickup_datetime'] = pd.to_datetime(car_df.date_time_of_pickup)
```

```python
## Modified - splitting datetime field into day,date, hour, etc.
car_df.loc[:, 'pickup_hour'] = car_df['pickup_datetime'].dt.hour
car_df.loc[:, 'pickup_weekday'] = car_df['pickup_datetime'].dt.day_name()
car_df.loc[:, 'pickup_date'] = car_df['pickup_datetime'].dt.day
car_df.loc[:, 'pickup_month'] = car_df['pickup_datetime'].dt.month
car_df.loc[:, 'pickup_day'] = car_df['pickup_datetime'].dt.dayofweek
```

```python
## Modified
## Convert date to date ordinal -- refer https://stackoverflow.com/questions/40217369/python-linear-regression-predict-by-date
import datetime as dt
car_df['pickup_datetime'] = pd.to_datetime(car_df['pickup_datetime'])
car_df['pickup_datetime']=car_df['pickup_datetime'].map(dt.datetime.toordinal)
```

```python
## Modified

#lets build our linear model
#independent varaibles
X=car_df.drop(["unique_id", "date_time_of_pickup", "pickup_weekday","amount"],axis=1)

## drop unique id - because it is not a feature, date_time_of_pickup - because it has been split into pickup_date and so on, pickup_weekday - because it is a string, amount - because it needs to be predicted
#the dependent variable
y=car_df[["amount"]]
```

```python
#split Xand y into trainning and test set in 10:30 ratio
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30)
```

```python
def ecdf(x):
    """Empirical cumulative distribution function of a variable"""
    # Sort in ascending order
    x = np.sort(x)
    n = len(x)
    
    # Go from 1/n to 1
    y = np.arange(1, n + 1, 1) / n
    
    return x, y
```

```python
xs, ys = ecdf(np.logspace(0, 2))
plt.plot(xs, ys, '.');
plt.ylabel('Percentile'); plt.title('ECDF');
```

```python
xs, ys = ecdf(mydata['amount'])
plt.figure(figsize = (8, 6))
plt.plot(xs, ys, '.')
plt.ylabel('Percentile'); plt.title('ECDF of Fare Amount'); plt.xlabel('Fare Amount ($)');
```
